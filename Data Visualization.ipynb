### Objective: To study the role of seniority in conversations, focusing on whether older participants tend to be conversation leaders. Additionally, to analyze if the age difference acts as a catalyst, making this trend more pronounced.

```python
import pandas as pd
import altair as alt
```

```python
# โหลดข้อมูล
df = pd.read_csv('/Users/kunanon/Desktop/STATxCUYear3/INFO VISUAL/Merge Data/combined_transcript_final_complete_cleaned.csv')
```

```python
# Percentage of Leadership Wins by Age Gap
# เตรียมข้อมูล
df_convo = df.drop_duplicates(subset=['conversation_id'])
df_filtered = df_convo[df_convo['question_leader'].isin(['Elder', 'Younger'])]

counts = (
    df_filtered
    .groupby(['age_diff', 'question_leader'])
    .size()
    .reset_index(name='count')
)

total_per_diff = counts.groupby('age_diff')['count'].transform('sum')
counts['percentage'] = (counts['count'] / total_per_diff) * 100
counts = counts.sort_values('age_diff')

# สร้างกราฟ
bar = alt.Chart(counts).mark_bar().encode(
    x=alt.X('age_diff:O',
            title='Age Difference (Years)',
            sort='ascending'),
    xOffset='question_leader:N',
    y=alt.Y('percentage:Q',
            title='Percentage (%)'),
    color=alt.Color(
        'question_leader:N',
        scale=alt.Scale(
            domain=['Elder', 'Younger'],
            range=["#F98F49", "#0B82FA"]
        ),
        title='Leader'
    )
).properties(
    width=700,
    height=400,
    title='Percentage of Leadership Wins by Age Gap'
)

Percentage_of_Leadershi_Wins_by_Age_Gap_chart = bar
Percentage_of_Leadershi_Wins_by_Age_Gap_chart
# save เป็น png
#final_chart.save('/Users/kunanon/Desktop/STATxCUYear3/INFO VISUAL/chart1.png')
```

```python
# Percentage of Speaking Time Wins by Age Gap
# Step 1: คำนวณ Time Leader
time_leader_map = {}

for conv_id, group in df.groupby('conversation_id'):
    if len(group) == 2:
        group = group.sort_values('age')
        younger = group.iloc[0]
        older = group.iloc[1]

        winner = 'Tie'

        if older['age'] > younger['age']:
            if older['total_speaking_time'] > younger['total_speaking_time']:
                winner = 'Elder'
            elif younger['total_speaking_time'] > older['total_speaking_time']:
                winner = 'Younger'
        else:
            winner = 'Same Age'

        time_leader_map[conv_id] = winner

df['time_leader'] = df['conversation_id'].map(time_leader_map)


# Step 2: เตรียมข้อมูลให้ 1 คู่ = 1 observation และตัด same age, tie
df_convo = df.drop_duplicates(subset=['conversation_id'])
df_filtered = df_convo[df_convo['time_leader'].isin(['Elder', 'Younger'])]

counts = (
    df_filtered
    .groupby(['age_diff', 'time_leader'])
    .size()
    .reset_index(name='count')
)

total_per_diff = counts.groupby('age_diff')['count'].transform('sum')
counts['percentage'] = (counts['count'] / total_per_diff) * 100
counts = counts.sort_values('age_diff')


# Step 3: Altair Visualization 
bar = alt.Chart(counts).mark_bar().encode(
    x=alt.X('age_diff:O',
            title='Age Difference (Years)',
            sort='ascending'),
    xOffset='time_leader:N',  
    y=alt.Y('percentage:Q',
            title='Percentage of Wins (%)'),
    color=alt.Color(
        'time_leader:N',
        scale=alt.Scale(
            domain=['Elder', 'Younger'],
            range=["#A50A7C", "#61F74E"]
        ),
        title='Who spoke more?'
    )
).properties(
    width=700,
    height=400,
    title='Percentage of Speaking Time Wins by Age Gap'
)


Percentage_of_Speaking_Time_Wins_by_Age_Gap_chart = bar 
Percentage_of_Speaking_Time_Wins_by_Age_Gap_chart

# save เป็น png
#Percentage_of_Speaking_Time_Wins_by_Age_Gap_chart.save('Percentage_of_Speaking_Time_Wins_by_Age_Gap_chart.png')
```

```python
# Percentage of Word Count Wins by Age Gap
# Step 1: คำนวณ Word Leader
word_leader_map = {}

for conv_id, group in df.groupby('conversation_id'):
    if len(group) == 2:
        group = group.sort_values('age')
        younger = group.iloc[0]
        older = group.iloc[1]

        winner = 'Tie'

        if older['age'] > younger['age']:
            if older['total_words'] > younger['total_words']:
                winner = 'Elder'
            elif younger['total_words'] > older['total_words']:
                winner = 'Younger'
        else:
            winner = 'Same Age'

        word_leader_map[conv_id] = winner

df['word_leader'] = df['conversation_id'].map(word_leader_map)

# Step 2: เตรียมข้อมูล
df_convo = df.drop_duplicates(subset=['conversation_id'])
df_filtered = df_convo[df_convo['word_leader'].isin(['Elder', 'Younger'])]

counts = (
    df_filtered
    .groupby(['age_diff', 'word_leader'])
    .size()
    .reset_index(name='count')
)

total_per_diff = counts.groupby('age_diff')['count'].transform('sum')
counts['percentage'] = (counts['count'] / total_per_diff) * 100
counts = counts.sort_values('age_diff')

# Step 3: Altair Visualization
bar = alt.Chart(counts).mark_bar().encode(
    x=alt.X('age_diff:O',
            title='Age Difference (Years)',
            sort='ascending'),
    xOffset='word_leader:N',  
    y=alt.Y('percentage:Q',
            title='Percentage of Wins (%)'),
    color=alt.Color(
        'word_leader:N',
        scale=alt.Scale(
            domain=['Elder', 'Younger'],
            range=["#363635", "#7DE8DB"]
        ),
        title='Who used more words?'
    ),
    
).properties(
    width=700,
    height=400,
    title='Percentage of Word Count Wins by Age Gap'
)

Percentage_of_Word_Count_Wins_by_Age_Gap_chart = bar
Percentage_of_Word_Count_Wins_by_Age_Gap_chart
# save เป็น png
#Percentage_of_Word_Count_Wins_by_Age_Gap_chart.save('Percentage_of_Word_Count_Wins_by_Age_Gap_chart.png')
```

```python
# Percentage of Perceived Leadership Wins by Age Gap
# Step 1: Data Prep
# ---------------------------------------------------------
df_convo = df.drop_duplicates(subset=['conversation_id'])

df_filtered = df_convo[
    df_convo['actual_leader'].isin(['Elder', 'Younger'])
]

counts = (
    df_filtered
    .groupby(['age_diff', 'actual_leader'])
    .size()
    .reset_index(name='count')
)

total_per_diff = counts.groupby('age_diff')['count'].transform('sum')
counts['percentage'] = (counts['count'] / total_per_diff) * 100
counts = counts.sort_values('age_diff')

# Step 2: Altair Visualization
bar = alt.Chart(counts).mark_bar().encode(
    x=alt.X('age_diff:O',
            title='Age Difference (Years)',
            sort='ascending'),
    xOffset='actual_leader:N',  
    y=alt.Y('percentage:Q',
            title='Percentage of Wins (%)'),
    color=alt.Color(
        'actual_leader:N',
        scale=alt.Scale(
            domain=['Elder', 'Younger'],
            range=["#587A9D", "#F9B311"]
        ),
        title='Who was voted as leader?'
    ),
).properties(
    width=700,
    height=400,
    title='Percentage of Perceived Leadership Wins by Age Gap'
)

Percentage_of_Perceived_Leadership_Wins_by_Age_Gap_chart = bar
Percentage_of_Perceived_Leadership_Wins_by_Age_Gap_chart

# save เป็น png
#Percentage_of_Perceived_Leadership_Wins_by_Age_Gap_chart.save('Percentage of Perceived Leadership Wins by Age Gap.png')
```
